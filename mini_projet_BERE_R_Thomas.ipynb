### 4. Opération sur le fichier sur les localités du Burkina Faso

```python
burkina = pd.read_csv('burkina_location.csv')
burkina.head()
```

```python
gounghin = burkina[burkina['location_name'].isin(['Gounghin'])]
gounghin
```

```python
## Export des données sur Gounghin
gounghin.to_csv('gounghin.csv',index=False)
```

```python
alphabet=list(string.ascii_uppercase)[0:16]
alphabet
```

```python
# Extraction la sous partie dont les premières lettres des noms des lieux
# sont compris entre 'A' et 'P'
A_to_P = burkina.loc[burkina['location_name'].str.startswith(tuple(alphabet))].sort_values(by='location_name')
A_to_P
```

```python
burkina.sort_values(by='lat').iloc[0,1] # Localité de la latitude minimale
```

```python
burkina.sort_values(by='long').iloc[0,1]  # Localité de la longitude minimale
```

```python
# lieux dont les coordonnées sont comprises ente (lat>=11 et long <=0.5)
filtre = burkina.loc[(burkina['lat']>=11) & (burkina['long']<=0.5)].sort_values(by=['lat','long'],ascending=[True,False])
filtre
```
